### 구두 과제 
 A라는 사람과 B 사람이 주사위 게임을 함. 
 randint 를 사용하여 누가 이겼는지 비겼는지 나타내는 코드를 짜시오. 
 동점이면 누군가 이길 때까지

```python
import random as rd

def make_dice_number(a,b):
    a = rd.randint(1,6)
    b = rd.randint(1,6)
    print("a = {}, b = {}".format(a,b))
    return a,b
    
def compare_dice(a,b):
    if a > b:
        print("winner = A")
    elif a < b:
        print("winner = B")
    else: # 동점일 경우
        a,b = make_dice_number(a,b)
        compare_dice(a,b)    

a=0
b=0
a,b = make_dice_number(a,b)
compare_dice(a,b)
```

### 의사 랜덤과 시드번호

```python
def pseudo_rand(x):
    a = 1103515245
    b = 12345
    m = 2 ** 31
    new_x = (a*x + b) % m
    return new_x

x = pseudo_rand(100)
print(x)
x = pseudo_rand(101)
print(x)
```
