### 2.	직각삼각형의 빗변을 이루는 선분의 양 끝점을 나타내는 두 좌표 (x1, y1), (x2, y2)를 입력받아 직각삼각형의 면적을 구하는 프로그램을 작성하여라. 이때 4개의 좌표값을 입력으로 받아 직각삼각형의 면적을 반환하는 area(x1, y1,x2, y2) 함수를 구현하시오.단, x2>x1이고 y2>y1이다.



```python
def tri_area(x1, y1, x2, y2):
        length = x2 - x1
        height = y2 - y1
        area = length * height / 2

        print("직각삼각형의 면적은 : {}".format(area))

x1 = int(input("x1 좌표를 입력하시오 :"))
y1 = int(input("y1 좌표를 입력하시오 :"))
x2 = int(input("x2 좌표를 입력하시오 :"))
y2 = int(input("y2 좌표를 입력하시오 :"))

tri_area(x1, y1, x2, y2)
    
```

### ***4.	두 개의 정수 a, b를 입력으로 받는다. 이때 a < b라는 조건이 항상 지켜진다고 하자. 여러분이 작성할 코드는 a에서 b까지(a, b 포함) 범위의 모든 정수의 배수이면서 가장 작은 값을 갖는 값을 찾는 것이다. 예를 들어 2, 4가 입력되면 2, 3, 4 모두의 배수이면서 가장 작은 값이 되는 12가 답이 된다. 12는 2의 배수, 3의 배수, 4의 배수이다. !



```python
# 이 문제는 해결방법을 찾지 못했습니다.

a = int(input("범위의 시작 정수 : "))
b = int(input("범위의 마지막 정수 : "))

result = 1 
for i in range (a, b+1):
    result *= i
    
print("{}에서 {}까지의 정수들의 최소 공배수는 : {}".format(a, b, result))
```

### 5.	분자가 1인 분수를 단위 분수라고 한다. 1보다 작은 소수값이 주어졌을 때 이 소수값에 가장 가까운 단위 분수를 찾는 함수 def unit_fraction(frac)을 구현하자. 예를 들어 0.27이 주어지면 1/3은 0.33333….이고 1/4는 0.25이며 1/5는 0.2이므로 1/4가 가장 가까운 단위분수이다. 이렇게 가장 가까운 단위분수를 찾아 출력하고 이 단위분수를 소수값으로 표현한 결과도 같이 출력하여라.!



```python
f = float(input("1보다 작고 0보다  큰 소수를 입력하세요: "))

def unit_fraction(frac): #가장 가까운 단위분수 찾기
    # 먼저 사잇값부터 찾자.
    # for range 문으로 무한루프를 돌리는 방법을 몰라서 while로 접근
    result = 0 # 결과를 담을 분모
    deno = 2 # 사잇값을 찾을 분모
    while True:
        if (1/deno) > f  :
            deno += 1
        else :
            break
    # 1/demo와 1/demo-1 사이에 어디에 더 가까운지 찾아야함
    if f-(1/deno) > (1/(deno-1))-f:
        result = deno-1
    else:
        result = deno
    print("가장 가까운 단위 분수는 1/{}이며, 이 값은 {}입니다.".format(result,1/result ))
        
unit_fraction(f)
```

### 6.	m과 n이라는 이름의 매개변수를 두개 입력받아서 이 두 값 중에서 큰 값을 반환하는 max2(m,n) 함수와 작은 값을 반환하는 min2(m,n) 함수를 구현하시오. 100과 200을 인자로 넣어서 호출하여 그 결과를 확인하시오.!



```python
m = int(input("숫자m을 입력하시오 : "))
n = int(input("숫자n을 입력하시오 : "))

def max2(m,n):
    if m > n:
        return m
    else:
        return n
    
def min2(m,n):
    if m < n:
        return m
    else:
        return n
    
max_mn = max2(m,n)
min_mn = min2(m,n)

print("{}과 {}중 큰 수는 : {}".format(m,n,max_mn))
print("{}과 {}중 작은 수는 : {}".format(m,n,min_mn))
```

### 7.	미국에서 주로 사용되는 단위인 마일 값을 국제 표준 단위인 킬로미터(km)값으로 변경하고자 한다. 마일 값을 매개변수로 넘겨받아 킬로미터로 반환하는 mile2km(m) 함수를 구현하고 이 함수를 호출하여 1에서 5마일을 킬로미터로 출력하시오. 이때 for - in range()를 사용하여 반복출력이 되도록 하시오.(단, 1 마일은 1.61 킬로미터로 정의하시오)

### 8.다음과 같이 사용자로부터 두점의 좌표 (x1, y1), (x2, y2)를 입력받아 두 점 사이의 거리를 출력하시오. 이를 위하여 distance(x1, y1, x2, y2) 함수를 구현하시오



```python
def distance(x1, y1, x2, y2):
    width = abs(x2-x1)
    height = abs(y2-y1)
    dis= ((width**2)+(height**2))**(1/2)
    print("두 점의 거리 : {:.1f}".format(dis))

x1 = int(input("x1의 좌표를 입력하시오 : "))
y1 = int(input("y1의 좌표를 입력하시오 : "))
x2 = int(input("x2의 좌표를 입력하시오 : "))
y2 = int(input("y2의 좌표를 입력하시오 : "))

distance(x1, y1, x2, y2)
```

### 9.	3개 숫자를 오름차순으로 정렬하여 출력하는 sort3() 함수를 작성하시오. 사용자로부터 3개의 숫자를 입력받고 이들 숫자를 오름차순으로 출력하는 테스트 프로그램을 작성하시오(힌트 : sort3() 함수의 매개변수로 입력된 정수는 리스트의 append() 메소드를 통해서 추가한 후 리스트의 sort() 메소드를 이용하여 정렬하시오. 정렬된 리스트를 a라 하면 a[0], a[1], a[2] 인덱스를 사용하여 출력하시오.



```python
def sort3():
    list_num = []
    print("세 수를 입력하세요 :")
    
    for i in range(3):
        temp = int(input(""))
        list_num.append(temp)
        
    list_num.sort()
    print("정렬된 리스트는 다음과 같습니다 :", end=' ')
    for i in range(len(list_num)):
        print(list_num[i], end=' ')

sort3()
```

### 10.	삼각형의 밑변(width), 높이(height)를 입력받아서 삼각형의 면적을 구하는 cal_area(width, height)라는 함수를 정의하여라. 사용자로부터 삼각형의 밑변과 높이를 입력받은 후, 이 함수를 이용하여 다음과 같은 결과를 얻는 프로그램을 작성하여라.



```python
def cal_area(width, height):
    area = width*height/2
    print("삼각형의 면적 : {:.1f}".format(area))
    
width = int(input("밑변을 입력하시오 : "))
height = int(input("높이를 입력하시오 : "))

cal_area(width, height)
```
