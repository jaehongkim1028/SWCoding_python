### 1.1.	로미오와 줄리엣 두 사람이 주사위를 던져서 높은 숫자가 나오면 이기는 게임을 만들어보자. 로미오와 줄리엣의 주사위는 모두 다음과 같이 random 모듈의 randrange( )를 통해서 생성한 난수를 바탕으로 한다. 출력 결과는 다음과 같이 로미오가 이기거나, 줄리엣이 이기거나, 비기는 결과가 나와야 한다!



```python
import random as rd

ro_num = rd.randrange(1,7)
print("로미오의 주사위 숫자는 {} 입니다.".format(ro_num))
ju_num = rd.randrange(1,7)
print("줄리엣의 주사위 숫자는 {} 입니다.".format(ju_num))

if (ro_num > ju_num):
    print("로미오가 이겼습니다.")
elif (ju_num > ro_num):
    print("줄리엣이 이겼습니다.")
else:
    print("비겼습니다.")
```

###  2.	랜덤 숫자 맞추기 게임 프로그램을 작성하라. 먼저 랜덤하게 1~20까지의 숫자 x를 하나 생성시키고 사용자는 숫자를 하나씩 입력하면서 생성된 숫자 x와 비교해 큰지 작은지를 보고 숫자를 맞춰가는 게임이다. 컴퓨터는 사용자가 입력한 숫자가 정답보다 큰지 작은지를 알려주어야 한다. 그리고 사용자가 시도를 할 때 마다 시도한 횟수가 n에 저장되도록 한다. 만일 입력한 숫자와 x가 일치하면 “정답입니다!”라는 메시지를 출력한다. 이때 3번 이하로 입력하여 숫자를 맞추면 “n번 만에 맞춘 당신은 천재!”, 3번 이상 6번 이하로 입력하여 숫자를 맞추면 “n번 만에 맞추셨네요. 잘 했어요^^”, 7번 이상으로 숫자를 입력하여 맞추면 “n번 만에 맞추다니 쩝쩝…”이라고 출력하라.!



```python
import random as rd

count = 0
num = rd.randrange(1,21)

while(True):
    count = count+1
    user_num = int(input("1~20까지의 숫자를 입력하세요 : "))
    if num ==  user_num:
        print("정답입니다!")
        print("{}번만에 숫자를 맞추셨네요. 잘했어요^^".format(count))
        break;
    elif user_num > num:
        print("{}보다 작습니다!".format(user_num))
    else:
        print("{}보다 큽니다!".format(user_num))
```

### 6.	춘향이와 몽룡이는 2019년 2월 24일 발렌타인데이에 연애를 시작했다. 오늘은 두사람의 연애 시작일로부터 며칠이 경과하였는가? Datetime 모듈을 사용하고 오늘날짜 date.today()를 통해 받아와서 오늘날짜, 연애를 시작한 날짜와 경과한 날짜, 기념일을 다음과 같이 출력하여라.!



```python
import datetime as dt

today = dt.date.today()
first_meet = dt.date(2019, 2, 24)

hundminusone = dt.timedelta(days=99)
hundred = dt.timedelta(days=100)

af_hund = first_meet + hundminusone
af_twoh = first_meet+ hundminusone + hundred
af_fiveh = first_meet+ hundminusone + hundred*4
af_thou = first_meet+ hundminusone + hundred*9

timegap = today - first_meet
print("오늘은 {}년 {}월 {}일입니다.".format(today.year, today.month, today.day))
print("춘향이와 몽룡이의 연애 시작일: {}년 {}월 {}일".format(first_meet.year, first_meet.month, first_meet.day))
print("연애 시작일로부터 경과한 날짜: {}일".format(timegap.days))
print("100일 기념일 : {}년 {}월 {}일".format(af_hund.year, af_hund.month, af_hund.day))
print("200일 기념일 : {}년 {}월 {}일".format(af_twoh.year, af_twoh.month, af_twoh.day))
print("500일 기념일 : {}년 {}월 {}일".format(af_fiveh.year, af_fiveh.month, af_fiveh.day))
print("1000일 기념일 : {}년 {}월 {}일".format(af_thou.year, af_thou.month, af_thou.day))
```

### 7.	1부터 1,000,000까지 정수의 합을 구하여 반환하는 함수 sum1to1000000()을 작성하고 함수를 100번 호출하여 함수 수행 시간을 다음과 같이 초단위로 구하여 출력하여라. (소수점 4자리까지 표현할 것.)



```python
import time as t
import datetime as dt

def sum1to1000000(sum):
    for i in range(1000000):
        sum = sum + i  
    return sum

sum = 0

# start
start_time = t.time()
#start_time = dt.datetime.now()

for i in range(100):
    sum = sum1to1000000(sum)

#end
finish_time = t.time()
#finish_time = dt.datetime.now()

timegap = finish_time - start_time

print("1에서 1,000,000까지의 합을 100번 반복해서 구하는 시간: {:6.4f}초".format(timegap))
    
```

### 8.	다음과 같은 함수를 작성하고 100번 반복하여 함수 수행 시간을 다음과 같이 초단위로 구하여 출력하여라.

(1) 1000 팩토리얼을 구하는 함수


(2) 1에서 1000까지의 모든 홀수를 세제곱하여 더한 것을 반환하는 함수


(3) 1도에서 360도까지의 sin값을 1도 단위로 구하여 그 합을 반환하는 함수


```python
import time as t
import math as m

# (1)
def fact_1000():
    fact = 1
    for i in range(1, 1001):
        fact *= i
    return fact
    
fact = 0
s_time = t.time()
for i in range(100):
    fact = fact_1000()
e_time = t.time()
print("1000!을 100번 반복해서 구하는 시간: {:6.4f}초".format(e_time - s_time))

# (2)
def odd_sum():
    sum = 0
    for i in range(1,1000,2):
        sum += i**3
    return sum

sum = 0
s_time = t.time()
for i in range(100):
    sum = odd_sum()
e_time = t.time()

print("1부터 1000까지 홀수의 세제곱 더하기 결과: {}".format(sum))
print("1부터 1000까지 홀수의 세제곱 더하기를 100번 반복하는 시간: {:6.4f}초".format(e_time - s_time))

# (3)
def find_sin_sum():
    sin_sum = 0
    for i in range(1,361):
        sin_sum += m.sin(m.radians(i))
    return sin_sum

s_time = t.time()
for i in range(100):
    sin_sum = find_sin_sum()
e_time = t.time()

print("1에서 360도까지 sin값의 합: {}".format(sin_sum))
print("1에서 360도까지 sin값 더하기를 100번 반복하는 시간: {:6.4f}초".format(e_time - s_time))
```
